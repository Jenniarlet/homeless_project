```python
import pandas as pd 
# path for the pop states excel
pop_states_path = '../data/raw/Est_popu_states_Census.xlsx'
```

```python
pop_states_df = pd.read_excel(pop_states_path, engine='openpyxl', sheet_name=list(range(2)))
```

```python
pop_states_total = pop_states_df[0].dropna().reset_index(drop=True)
```

```python
columns = pop_states_total.columns
pop_states_total = pop_states_total.rename(columns={columns[0]: 'geographic_area', columns[1]: 'base_estimate_2020',
                                 columns[2]: 'popu_estimate_2020', columns[3]: 'popu_estimate_2021'})
```

```python
pop_states_total['geographic_area'] = pop_states_total.apply(lambda row: row.geographic_area.replace('.', ''), axis=1)
```

```python
pop_states_total.to_csv('../data/processed/geo_area_popu_estimate.csv', index=False)
```
