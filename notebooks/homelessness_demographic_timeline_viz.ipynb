```python
merged_path = '../../data/processed/pit_hic_with_common_coc.csv'
```

```python
import pandas as pd
merged = pd.read_csv(merged_path)
```

```python
import seaborn as sns
from matplotlib import pyplot as plt
def create_bplot(indiv_fields, total, title, palette='dark'):
  filt_merged = merged[indiv_fields + ['Year']]
  filt_merged = filt_merged[filt_merged[indiv_fields[1]].notnull()]
  filt_annual = filt_merged.groupby('Year').sum().reset_index()
  # filt_annual = filt_annual.rename(columns={total: total + ' - Total'})
  df = pd.melt(filt_annual, id_vars='Year', var_name="Homelessness Status", value_name='Homelessness Count')
  sns.set_style("darkgrid")
  #  {'grid.linestyle': '--'}
  # ax.ticklabel_format(style='sci', axis='both')
  title = title + f' ({df.Year.min()} - {df.Year.max()})'
  g = sns.catplot(x='Year', y='Homelessness Count', hue='Homelessness Status', data=df, kind='bar', palette=palette, aspect=2, height=6, orient="v").set(title=title)
  g.despine(left=True)
  # g.set_axis_labels("", "Body mass (g)")
  g.legend.set_title("")

  # plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
  
  if total:
    filt = merged[[total, 'Year']]
    new_total = total + ' - Total'
    filt = filt.rename(columns={total:new_total})
    filt.groupby('Year').sum().plot()
    # plt.ticklabel_format(style='sci', axis='y', scilimits=(0,0))
    plt.ylabel('Homelessness Count')
  
    # filt_merged[[new_total, 'Year']].groupby('Year').sum().plot()
```

```python
age = ['Unsheltered Homeless - Under 18',
 'Unsheltered Homeless - Age 18 to 24',
 'Unsheltered Homeless - Over 24']

create_bplot(age, 'Unsheltered Homeless', 'Unsheltered Homeless by age', palette='tab10')
```

```python
gender = ['Unsheltered Homeless - Female',
 'Unsheltered Homeless - Male']

gender2 = ['Unsheltered Homeless - Transgender',
 'Unsheltered Homeless - Gender Non-Conforming']
create_bplot(gender + gender2, None, 'Unsheltered Homeless by gender')

create_bplot(gender, None, 'Unsheltered Homeless by gender')
create_bplot(gender2, 'Unsheltered Homeless', 'Unsheltered Homeless by gender', palette='tab10')
```

```python
race = [ 'Unsheltered Homeless - Non-Hispanic/Non-Latino',
 'Unsheltered Homeless - Hispanic/Latino',
 'Unsheltered Homeless - White',
 'Unsheltered Homeless - Black or African American',
 'Unsheltered Homeless - Asian',
 'Unsheltered Homeless - American Indian or Alaska Native',
 'Unsheltered Homeless - Native Hawaiian or Other Pacific Islander',
 'Unsheltered Homeless - Multiple Races']
create_bplot(race, None, 'Unsheltered Homeless by race', palette='tab10')
create_bplot(race[:4], None, 'Unsheltered Homeless by race')
create_bplot(race[4:], 'Unsheltered Homeless', 'Unsheltered Homeless by race', palette='husl')
```
